# Gaussian distribution

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from scipy.stats import norm
```

```python
samples=1000
theta = np.random.uniform(0,2*np.pi,samples)
r = np.random.uniform(1,2,samples)
x = r * np.cos(theta)
y = r * np.sin(theta)
```

```python
plt.scatter(x,y)
```

```python
sns.distplot(x,fit=norm)
```

```python
def ring(samples):
    theta = 360 * np.random.uniform(0,2*np.pi,samples)
    r = np.random.uniform(1,2,samples)
    return  r * np.cos(theta)
```

```python
x1=ring(1000)
x2=ring(1000)
x3=ring(1000)
x4=ring(1000)
x5=ring(1000)
x6=ring(1000)
```

```python
sns.distplot(x1+x2+x3+x4,fit=norm)
```
